### Load requried libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Download data

We are using UCI ml iris data

```python
# 1. Download a dataset from the web (UCI ML Repository - Iris dataset)
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
column_names = ["SepalLength", "SepalWidth", "PetalLength", "PetalWidth", "Species"]
iris_data = pd.read_csv(url, header=None, names=column_names)
```

```python
# 2. Print out descriptive statistics
# Selecting a few quantitative and categorical variables for demonstration
quantitative_vars = ["SepalLength", "PetalLength", "PetalWidth"]
categorical_var = "Species"
```

```python
# Descriptive statistics for quantitative variables
print(iris_data[quantitative_vars].describe())

# Descriptive statistics for categorical variable
print(iris_data[categorical_var].value_counts())
```

```python
# 3. Transform at least one variable (e.g., square root transformation of SepalLength)
iris_data["Transformed_SepalLength"] = np.sqrt(iris_data["SepalLength"])
iris_data.head()
```

```python
# 4. Plot at least one quantitative variable and one scatterplot
# Histogram for SepalLength
plt.figure(figsize=(8, 6))
sns.histplot(iris_data["SepalLength"], bins=20, kde=True, color="skyblue")
plt.title("Histogram of SepalLength")
plt.xlabel("SepalLength")
plt.ylabel("Frequency")
plt.show()

# Scatterplot for SepalLength vs. PetalLength
plt.figure(figsize=(8, 6))
sns.scatterplot(x="SepalLength", y="PetalLength", hue="Species", data=iris_data)
plt.title("Scatterplot of SepalLength vs. PetalLength")
plt.xlabel("SepalLength")
plt.ylabel("PetalLength")
plt.show()
```
